```python
import requests
import string
import json
from xml.etree.ElementTree import Element, SubElement, Comment, tostring
```

```python
def query_mms(website, term):
    headers = {
        "Accept": "application/json",
        "Accept-Language": "en",
        "api-version": "v2"
    }

    payload = {
        "q": term,
        "Flexisearch": "true"
    }


    r = requests.get(
        '{}/icd/release/11/2019-04/mms/search'.format(website),
        headers=headers,
        params=payload
    )
    
    return r
```

```python
def query_by_alphabet(website):
    return query_by_terms(website, string.ascii_lowercase)
```

```python
def query_by_terms(website, terms):
    results = {}
    for alpha in terms:
        resp = query_mms(website, alpha)
        resp_data = resp.json()
        results.update({item['theCode']:item for item in resp_data['destinationEntities']})
    return results
```
